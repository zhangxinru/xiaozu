```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
#读取训练集
train = pd.read_csv('/Users/Administrator/Desktop/train.csv')
#读取测试集
test = pd.read_csv('/Users/Administrator/Desktop/test.csv')
#features选取
X_train1 = train[['I1','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I12','I13']]
X_test1 = test[['I1','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I12','I13']]
X_train2=train[['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11']] 
X_test2 = test[['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11']]
y_train = train['Label']

X_train1['I5'].fillna(X_train1['I5'].mean(),inplace = True)
X_train1['I6'].fillna(X_train1['I6'].mean(),inplace = True)
X_train1['I7'].fillna(X_train1['I7'].mean(),inplace = True)
X_train1['I8'].fillna(X_train1['I8'].mean(),inplace = True)
X_train1['I9'].fillna(X_train1['I9'].mean(),inplace = True)
X_train1['I11'].fillna(X_train1['I11'].mean(),inplace = True) 
X_test1['I5'].fillna(X_test1['I5'].mean(),inplace = True)
X_test1['I6'].fillna(X_test1['I6'].mean(),inplace = True)
X_test1['I7'].fillna(X_test1['I7'].mean(),inplace = True)
X_test1['I8'].fillna(X_test1['I8'].mean(),inplace = True)
X_test1['I9'].fillna(X_test1['I9'].mean(),inplace = True)
X_test1['I11'].fillna(X_test1['I11'].mean(),inplace = True)

def set_missing_I1s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I1_df = df[['I1','I2','I5','I6','I7','I8','I9','I11']]  
    known_I1 = I1_df[I1_df.I1.notnull()].as_matrix() 
    unknown_I1 = I1_df[I1_df.I1.isnull()].as_matrix()
    # y即目标
    y = known_I1[:, 0] 
    # X即特征属性值 
    X = known_I1[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI1s = rfr.predict(unknown_I1[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I1.isnull()), 'I1' ] = predictedI1s 
    return df, rfr 
X_train1 ,rfr = set_missing_I1s(X_train1)
tmp_df = X_test1[['I1','I2','I5','I6','I7','I8','I9','I11']]
null_I1 = tmp_df[X_test1.I1.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I1[:, 1:] 
predictedI1s = rfr.predict(X) 
X_test1.loc[ (X_test1.I1.isnull()), 'I1' ] = predictedI1s

def set_missing_I3s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I3_df = df[['I3','I2','I1','I5','I6','I7','I8','I9','I11']]  
    known_I3 = I3_df[I3_df.I3.notnull()].as_matrix() 
    unknown_I3 = I3_df[I3_df.I3.isnull()].as_matrix()
    # y即目标
    y = known_I3[:, 0] 
    # X即特征属性值 
    X = known_I3[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI3s = rfr.predict(unknown_I3[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I3.isnull()), 'I3' ] = predictedI3s 
    return df, rfr 
X_train1 ,rfr = set_missing_I3s(X_train1)
tmp_df = X_test1[['I3','I2','I1','I5','I6','I7','I8','I9','I11']] 
null_I3 = tmp_df[X_test1.I3.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I3[:, 1:] 
predictedI3s = rfr.predict(X) 
X_test1.loc[ (X_test1.I3.isnull()), 'I3' ] = predictedI3s  
 
def set_missing_I4s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I4_df = df[['I4','I2','I3','I1','I5','I6','I7','I8','I9','I11']]  
    known_I4 = I4_df[I4_df.I4.notnull()].as_matrix() 
    unknown_I4 = I4_df[I4_df.I4.isnull()].as_matrix()
    # y即目标
    y = known_I4[:, 0] 
    # X即特征属性值 
    X = known_I4[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI4s = rfr.predict(unknown_I4[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I4.isnull()), 'I4' ] = predictedI4s 
    return df, rfr 
X_train1 ,rfr = set_missing_I4s(X_train1)
tmp_df = X_test1[['I4','I2','I3','I1','I5','I6','I7','I8','I9','I11']] 
null_I4 = tmp_df[X_test1.I4.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I4[:, 1:] 
predictedI4s = rfr.predict(X) 
X_test1.loc[ (X_test1.I4.isnull()), 'I4' ] = predictedI4s 



def set_missing_I10s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I10_df = df[['I10','I2','I3','I4','I5','I6','I7','I8','I9','I1','I11']]  
    known_I10 = I10_df[I10_df.I10.notnull()].as_matrix() 
    unknown_I10 = I10_df[I10_df.I10.isnull()].as_matrix()
    # y即目标
    y = known_I10[:, 0] 
    # X即特征属性值 
    X = known_I10[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI10s = rfr.predict(unknown_I10[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I10.isnull()), 'I10' ] = predictedI10s 
    return df, rfr 
X_train1 ,rfr = set_missing_I10s(X_train1)
tmp_df = X_test1[['I10','I2','I3','I4','I5','I6','I7','I8','I9','I1','I11']] 
null_I10 = tmp_df[X_test1.I10.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I10[:, 1:] 
predictedI10s = rfr.predict(X) 
X_test1.loc[ (X_test1.I10.isnull()), 'I10' ] = predictedI10s 

def set_missing_I12s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I12_df = df[['I12','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I1']]  
    known_I12 = I12_df[I12_df.I12.notnull()].as_matrix() 
    unknown_I12 = I12_df[I12_df.I12.isnull()].as_matrix()
    # y即目标
    y = known_I12[:, 0] 
    # X即特征属性值 
    X = known_I12[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI12s = rfr.predict(unknown_I12[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I12.isnull()), 'I12' ] = predictedI12s 
    return df, rfr 
X_train1 ,rfr = set_missing_I12s(X_train1)
tmp_df = X_test1[['I12','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I1']] 
null_I12 = tmp_df[X_test1.I12.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I12[:, 1:] 
predictedI12s = rfr.predict(X) 
X_test1.loc[ (X_test1.I12.isnull()), 'I12' ] = predictedI12s 

def set_missing_I13s(df): # 把已有的数值型特征取出来丢进Random Forest Regressor中 
    I13_df = df[['I13','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I12','I1']]  
    known_I13 = I13_df[I13_df.I13.notnull()].as_matrix() 
    unknown_I13 = I13_df[I13_df.I13.isnull()].as_matrix()
    # y即目标
    y = known_I13[:, 0] 
    # X即特征属性值 
    X = known_I13[:, 1:] 
    # fit到RandomForestRegressor之中 
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1) 
    rfr.fit(X, y) 
    # 用得到的模型进行未知结果预测 
    predictedI13s = rfr.predict(unknown_I13[:, 1::]) 
    # 用得到的预测结果填补原缺失数据 
    df.loc[ (df.I13.isnull()), 'I13' ] = predictedI13s 
    return df, rfr 
X_train1 ,rfr = set_missing_I13s(X_train1)
tmp_df = X_test1[['I13','I2','I3','I4','I5','I6','I7','I8','I9','I10','I11','I12','I1']] 
null_I13 = tmp_df[X_test1.I13.isnull()].as_matrix() # 根据特征属性X预测并补上 
X = null_I13[:, 1:] 
predictedI13s = rfr.predict(X) 
X_test1.loc[ (X_test1.I13.isnull()), 'I13' ] = predictedI13s 
X_train=X_train1.join(X_train2)
X_test=X_test1.join(X_test2)
#DictVectorizer进行特征提取
dict_vec = DictVectorizer(sparse=False)
X_train = dict_vec.fit_transform(X_train.to_dict(orient='record'))
X_test = dict_vec.transform(X_test.to_dict(orient='record'))
#模型选择XGB
xgb_model = xgb.XGBClassifier()
 
#设置参数
params = dict(booster='gbtree',
              objective='multi:softmax',
              num_class=2,
              learning_rate=0.1,
              max_depth=2,
              silent=0,)
# 设置迭代次数
plst = list(params.items())
num_rounds = 2000
 
# sklearn.cross_validation进行训练数据集划分，训练集和交叉验证集比例
train_x, val_X, train_y, val_y = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
 
# xgb矩阵赋值
xgb_val = xgb.DMatrix(val_X, label=val_y)
xgb_train = xgb.DMatrix(train_x, label=train_y)
xgb_test = xgb.DMatrix(X_test)
 
#watchlist 方便查看运行情况
watchlist = [(xgb_train, 'train'), (xgb_val, 'val')]
 
# training model
# early_stopping_rounds 当设置的迭代次数较大时，early_stopping_rounds 可在一定的迭代次数内准确率没有提升就停止训练
model = xgb.train(plst, xgb_train, num_rounds, watchlist, early_stopping_rounds=100)
 
#测试集合预测值
preds = model.predict(xgb_test, ntree_limit=model.best_ntree_limit)
#结果输出
np.savetxt('/Users/Administrator/Desktop/CRITEO2.csv', np.c_[range(1, len(X_test) + 1), preds], delimiter=',', header='Label', comments='', fmt='%d')
```
